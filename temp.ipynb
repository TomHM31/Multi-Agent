```python
# Import the statcast function from pybaseball
from pybaseball import statcast

# Retrieve all Statcast data for the Colorado Rockies in the 2019 season
data = statcast(start_dt='2019-03-28', end_dt='2019-09-29', team='COL')
# Filter the data to only home runs
hr_data = data[data['events'] == 'home_run']
# Group the home run data by batter and count the number of home runs for each
hr_by_batter = hr_data.groupby('batter').size()
# Find the batter with the most home runs
hr_leader = hr_by_batter.idxmax()
num_hr = hr_by_batter.max()
from pybaseball import playerid_reverse_lookup

print(playerid_reverse_lookup.__doc__)
# Look up the player name for the home run leader 
from pybaseball import playerid_reverse_lookup

player_info = playerid_reverse_lookup([hr_leader], key_type='mlbam')
player_name = player_info.loc[0, 'name_first'] + ' ' + player_info.loc[0, 'name_last']
# Print out the home run leader and their home run total
print(f"{player_name} led the Colorado Rockies with {num_hr} home runs in the 2019 season.")
```
